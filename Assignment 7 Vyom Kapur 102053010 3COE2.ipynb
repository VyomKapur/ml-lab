```python
x = pd.DataFrame(boston["data"])
x.columns = boston["feature_names"]
y = pd.DataFrame(boston["target"])
```

```python
boston_data = pd.concat([x, y])
boston_data.corr()
```

```python
scaler_1 = StandardScaler()
x = scaler_1.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
kernel = ['linear', 'rbf', 'poly']
kernel_performance = []
for i in kernel:
    c_score = []
    for c in range(1, 100):
        model = SVR(C=c)
        model.fit(x_train, y_train)
        c_score.append((c,model.score(x_test, y_test)))
    kernel_performance.append((i, max(c_score, key=lambda item: item[1])))
```

```python
print(kernel_performance)
```

### Q3

```python
mall.corr()
```

```python
le = LabelEncoder()
mall["Gender"] = le.fit_transform(mall["Gender"])
```

```python
mall.drop("CustomerID", axis=1, inplace=True)
```

```python
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit(mall)
    wcss.append(km.inertia_)
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,11),wcss)
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
km1=KMeans(n_clusters=5)
km1.fit(mall)
y=km1.predict(mall)
mall["label"] = y
```

### Q5

```python
diabetes
```

```python
x = diabetes.iloc[:,0:8]
y = diabetes.iloc[:,8]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
MLP = MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=1000)
MLP.fit(x_train, y_train)
```

```python
y_pred = MLP.predict(x_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
MLP.score(x_test, y_test)
```
